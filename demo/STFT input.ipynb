```python
import scipy.io as io
import numpy as np
import utlis.utils as utils
#load time series data
x_test_series_mat = io.loadmat('./data/testset/x_test_series.mat')
x_test_series = x_test_series_mat['x_test_series_raw']
# STFT
x_test_stft = utils.batch_stft(x_test_series)
#load labels
#earthqauke [1,0,0,0]
#quake [0,1,0,0]
#rockfall [0,0,1,0]
#other noise [0,0,0,1]
y_test = np.load('./data/testset/y_test.npy')
```

```python
# Load pre-trained model
from tensorflow.keras import models
model=models.load_model('./models/multi_channel/stft_multi.h5')
# Prediction
classes = model.predict(x_test_stft, batch_size=256)
```

```python
# Results
utils.resultshow(classes,y_test)
```
