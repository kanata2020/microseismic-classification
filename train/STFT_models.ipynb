```python
#CNN model
from tensorflow.keras import Sequential, optimizers
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Dense, Flatten,GRU, Bidirectional, LSTM, BatchNormalization
from tensorflow.keras.models import load_model
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau

# model building
model = Sequential()
model.add(Conv2D(64, kernel_size=(3,3),strides=(1,1), activation='relu',input_shape=(128,75,6), name='conv1'))
model.add(MaxPooling2D(pool_size=(2,2),strides=None))
model.add(Conv2D(128, kernel_size=(3,3),strides=(1,1), activation='relu',name='conv2'))
model.add(MaxPooling2D(pool_size=(2,2),strides=None))
model.add(Conv2D(256, kernel_size=(3,3),strides=(1,1), activation='relu',name='conv3'))
model.add(Conv2D(256, kernel_size=(3,3),strides=(1,1), activation='relu',name='conv4'))
model.add(MaxPooling2D(pool_size=(2,2),strides=None))
model.add(Conv2D(512, kernel_size=(3,3),strides=(1,1), activation='relu',name='conv5'))
model.add(Conv2D(512, kernel_size=(3,3),strides=(1,1), activation='relu',name='conv6'))
model.add(MaxPooling2D(pool_size=(2,2),strides=None))
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(4, activation='softmax'))

#learning rate fitting
reduce_lr = ReduceLROnPlateau(monitor='loss', patience=5, factor=0.9, mode='auto')

model.compile(optimizer=optimizers.Adam(lr=0.0007, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False), loss='categorical_crossentropy',metrics=['accuracy'])
model.summary()

hist=model.fit(x_train,y_train,epochs=100,batch_size=64,validation_data=(x_valid,y_valid),callbacks=[reduce_lr])
```
